```python
import pandas as pd
import numpy as np
```

```python
# Load the input table
df = pd.read_csv("../docs/extended/extended_data_table_3.tsv", sep="\t")

print(f"Total number of rows: {len(df)}")
```

```python
SEED = 42  # a fixed integer 
np.random.seed(SEED)
```

```python
sample_fraction = 0.10
sample_size = int(len(df) * sample_fraction)

sampled_df = df.sample(
    n=sample_size,
    random_state=SEED
)

print(f"Selected {len(sampled_df)} rows ({sample_fraction*100:.0f}%)")
```

```python
sampled_df = sampled_df.sort_index()
```

```python
sampled_df.to_csv("../docs/extended/extended_data_table_3_random_10_percent.tsv",
    sep="\t",
    index=False
)
```
